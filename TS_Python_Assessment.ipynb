```python
# 1.palindrome -integer way
num = 121
temp = num
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

print("Palindrome" if temp == rev else "Not Palindrome")
```

```python
# 2.palindrome -string way
s = "madam"
print("Palindrome" if s == s[::-1] else "Not Palindrome")
```

```python
# 3.Fibbonacci series
n = 10
a, b = 0, 1

for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# 4.Armstrong number
num = 153
temp = num
digits = len(str(num))
total = 0

while num > 0:
    d = num % 10
    total += d ** digits
    num //= 10

print("Armstrong" if total == temp else "Not Armstrong")
```

```python
# 5.Reverse a string in 3 ways
s = "python"

# Method 1
print(s[::-1])

# Method 2
print("".join(reversed(s)))

# Method 3
rev = ""
for ch in s:
    rev = ch + rev
print(rev)
```

```python
# 7.Anagram
s1 = "listen"
s2 = "silent"

print("Anagram" if sorted(s1) == sorted(s2) else "Not Anagram")
```

```python
# 8.Count letter frequency in a string.
s = "programming"
freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

```python
# 9.Count word frequency in a string or sentence.
sentence = "hello world hello python"
words = sentence.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)
```

```python
# 10.How to create nested dictionary and if it contains list as a value how to append element and modify and remove from that list.
data = {
    "student": {
        "name": "Karthik",
        "skills": ["Python", "Java"]
    }
}

# Append
data["student"]["skills"].append("SQL")

# Modify
data["student"]["skills"][1] = "Spring"

# Remove
data["student"]["skills"].remove("Python")

print(data)
```

```python
# 11.Factorial
n = 5
fact = 1

for i in range(1, n + 1):
    fact *= i

print(fact)
```

```python
# 12.Prime number
num = 29

if num < 2:
    print("Not Prime")
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

```python
# 17.Find most frequent 2 words.
from collections import Counter

sentence = "hello world hello python world hello"
words = sentence.split()

print(Counter(words).most_common(2))
```

```python
# 18.sorting a random integer+string list in acceding order and descending order example input mixed_list = [9, 'c', 2, 'a', 5, 'b']
mixed_list = [9, 'c', 2, 'a', 5, 'b']

# Ascending
asc = sorted(mixed_list, key=lambda x: (str(type(x)), x))
print(asc)

# Descending
desc = sorted(mixed_list, key=lambda x: (str(type(x)), x), reverse=True)
print(desc)
```

```python
# 19.sort a dictionary based on its  keys or sort a dictionary based on its values
data = {"b": 3, "a": 1, "c": 2}

# By Keys
print(dict(sorted(data.items())))

# By Values
print(dict(sorted(data.items(), key=lambda x: x[1])))
```

```python
# 20.LCM and HCF (GCD)
import math

a, b = 12, 18

hcf = math.gcd(a, b)
lcm = (a * b) // hcf

print("HCF:", hcf)
print("LCM:", lcm)
```
